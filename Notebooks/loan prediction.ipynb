**Data Preprocessing**

```python
from sklearn.model_selection import train_test_split
```

```python
y=df['Loan_Status'].map({'Y': 1, 'N': 0})
x=df.drop(['Loan_Status','Loan_ID'], axis=1)
x=pd.get_dummies(x)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

**Training**

```python
!pip install catboost
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score
```

```python
voting=[]
```

**1. CatBoost Classifier**

```python
model = CatBoostClassifier()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
voting.append(y_pred)
```

**2. XGBoost Classifier**

```python
model = XGBClassifier()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
voting.append(y_pred)
```

**3. Light GBM Classifier**

```python
model = LGBMClassifier()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
voting.append(y_pred)
```

```python
y_pred=[]
for i in range(len(voting[0])):
    y_pred.append(max(set([voting[0][i],voting[1][i],voting[2][i]]), key=[voting[0][i],voting[1][i],voting[2][i]].count))
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
model.booster_.save_model('model.txt')
```

```python
import pickle
import numpy as np
import pandas as pd

# Save the LightGBM model to a .pkl file (existing code)
with open('lgbm_model.pkl', 'wb') as file:
    pickle.dump(model, file)
print("Model saved successfully to lgbm_model.pkl")

# Install onnxmltools if not already installed
!pip install onnxmltools onnx onnxruntime

from onnxmltools.convert import convert_lightgbm
from onnxmltools.convert.common.data_types import FloatTensorType # Corrected import

# Create a dummy input for ONNX conversion. onnxmltools requires a numpy array.
# The input should have the same number of features as the training data.
dummy_input = x_train.iloc[:1].values.astype(np.float32)

# Convert the LightGBM model to ONNX format using onnxmltools
onx_model = convert_lightgbm(model, initial_types=[('input', FloatTensorType(list(dummy_input.shape)))])

# Save the ONNX model to a file
with open("lgbm_model.onnx", "wb") as f:
    f.write(onx_model.SerializeToString())
print("Model saved successfully to lgbm_model.onnx")
```
